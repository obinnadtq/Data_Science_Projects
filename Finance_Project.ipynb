# Finance Data Project

Import libraries

```python
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import datetime
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
start = datetime.datetime(2006, 1, 1)         #start date for data
end = datetime.datetime(2016,1,1)             #end date for data
```

```python
df = pd.read_pickle('all_banks')            #read file containing data
```

```python
#Bank of America
BAC = df['BAC']

#CitiGroup
C = df['C']

#Goldman Sachs
GS = df['GS']

#JPMorgan Chase
JPM = df['JPM']

#Morgan Stanley
MS = df['MS']

#Wells Fargo
WFC = df['WFC']

```

```python
tickers = ['BAC','C','GS','JPM','MS','WFC']   # list of ticker symbols
```

```python
bank_stocks = pd.concat([BAC,C,GS,JPM,MS,WFC],axis = 1,keys = tickers)   #concatenate data
```

```python
bank_stocks.columns.names = ['Bank Ticker','Stock Info']     #setting column name levels
```

```python
bank_stocks.head()
```

```python
bank_stocks.xs(key='Close',axis = 1,level = 'Stock Info').max()   #max close price for each of the bank's stock
```

```python
returns = pd.DataFrame()   #create an empty DataFrame
```

```python
for tick in tickers:
    returns[tick+' Return'] = bank_stocks[tick]['Close'].pct_change()
```

```python
returns.head()
```

```python
sns.pairplot(returns[1:])
```
