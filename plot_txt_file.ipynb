```python
import matplotlib.pyplot as plt
import numpy as np
import glob
%matplotlib inline
```

```python
data = sorted(glob.glob('rmsd*.txt'))
up_data = []
for file in data:
    new_data = np.loadtxt(file)
    up_data.append(new_data)
fig,axes = plt.subplots(nrows=2,ncols=5,figsize = (40,10))
for i in range(5):
    axes[0,i].plot(up_data[i][:,0],up_data[i][:,1])
    axes[0,i].set_xlabel('time(ns)')
    axes[0,i].set_ylabel('RMSD(A)')
    axes[0,i].set_title('RMSD {}'.format(i+1))
for i in range(5):
    axes[1,i].plot(up_data[5+i][:,0],up_data[5+i][:,1])
    axes[1,i].set_xlabel('time(ns)')
    axes[1,i].set_ylabel('RMSD(A)')
    axes[1,i].set_title('RMSD {}'.format(5+i+1))
plt.tight_layout()
```
